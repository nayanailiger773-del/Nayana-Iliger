```python
#function to get integer label from given label
def getLabel(name):
    index = -1
    for i in range(len(labels)):
        if labels[i] == name:
            index = i
            break
    return index
```

```python
import os
import numpy as np
import cv2

# Initialize arrays
X = []
Y = []

# Path to dataset folder
path = r"C:\Users\91962\Desktop\DDSM Dataset"

# Dummy label function (replace with your own)
def getLabel(name):
    return name  # Modify as per requirement

# Load saved dataset if exists
if os.path.exists("model/X.txt.npy"):
    X = np.load("model/X.txt.npy")
    Y = np.load("model/Y.txt.npy")
else:
    for root, dirs, files in os.walk(path):  # loop through dataset folder
        for file in files:
            if "Thumbs.db" in file or "mask" in file:
                continue

            img_path = os.path.join(root, file)
            img = cv2.imread(img_path)

            if img is not None:
                img = cv2.resize(img, (32, 32))  
                X.append(img)

                label_name = os.path.basename(root)
                label = getLabel(label_name)
                Y.append(label)

    X = np.asarray(X)
    Y = np.asarray(Y)

    os.makedirs("model", exist_ok=True)
    np.save("model/X.txt.npy", X)
    np.save("model/Y.txt.npy", Y)

print("Dataset Features Extraction & Loading Completed")
print(f"Total images loaded = {X.shape[0]}")
```
